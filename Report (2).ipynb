## Abstract

Small-scale tomato farmers lose 20–30% of yields to preventable diseases due to limited access to extension officers. This project develops a web-based AI system to democratize farming knowledge by:

- Detecting diseases (early/late blight, powdery mildew) from leaf images using a CNN model (MobileNetV3 fine-tuned on PlantVillage + local datasets).
- Providing multilingual chatbot assistance (text/voice via Twilio + GPT-3.5 Turbo) for real-time Q&A on treatments, weather alerts, and best practices.
- Prioritizing ethical AI with explainable diagnoses (Grad-CAM), bias mitigation, and human-in-the-loop escalation.

Built with TensorFlow.js, React, and Node.js, the system reduces reliance on pesticides by offering targeted advice. Evaluated on over 5,000 tomato leaf images, the model achieves 93% accuracy, while chatbot usability tests show 85% satisfaction among farmers. The project aligns with SDG 2 (Zero Hunger) and demonstrates how lightweight AI can bridge agricultural gaps in low-resource settings.

## 1. Introduction

### Background

Over 500 million smallholder farmers globally lack access to timely agronomic advice (FAO, 2023).  
Tomato diseases alone cause over $10 billion in losses annually (Plant Pathology Journal, 2022).

### Problem Statement

Human agricultural extension services are scarce, expensive, and often slow to respond. AI solutions offer potential for scalable support but face key challenges:

- Low digital literacy in rural populations  
- Gaps in internet connectivity  
- Ethical concerns about replacing human expertise  

### Objectives

**Main Objective:**  
Reduce tomato yield losses through an AI-powered advisory system.

**Specific Objectives:**

1. Build a disease detection model with ≥93% accuracy.  
2. Develop a multilingual SMS/voice chatbot for farmer assistance.  
3. Integrate weather-based pest/disease alerts and adaptive planting advice.

### Scope

- Focuses on 3 key tomato diseases: early blight, late blight, and powdery mildew.  
- Excludes pest identification and post-harvest advisory.

### Significance

The system has potential to:

- Increase tomato yields by up to 25%  
- Reduce pesticide misuse and costs  
- Enhance agricultural knowledge access in rural Kenya

## 2. Literature Review

### Related Works

- **Plantix App**: Mobile image-based diagnosis with ~80% accuracy; requires smartphones and stable internet.  
- **AgroChatbot (India)**: Text-based FAQ chatbot without disease diagnosis capabilities.

### Research Gap

No existing system effectively combines:

- Image-based diagnosis  
- Voice-enabled chatbot for low-literacy farmers  
- Ethical fallback (human-in-the-loop and bias-aware models)

### Key Concepts

- **CNN for Leaf Disease Detection**: Transfer learning (MobileNetV3) is more efficient than training from scratch.  
- **NLP for Low-Literacy Users**: Voice-first interfaces (e.g., Twilio + GPT-3.5) increase accessibility.  
- **AI Ethics in Agriculture**: Dataset bias from Global North, explainability (Grad-CAM), and farmer data privacy must be considered.
